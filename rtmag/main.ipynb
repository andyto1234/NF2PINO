```python
from pathlib import Path
from urllib.request import urlretrieve
import torch

model_files = {
    "args.npy": "https://huggingface.co/mingyujeon/rtmag-models/resolve/main/isee/args.npy",
    "best_model.pt": "https://huggingface.co/mingyujeon/rtmag-models/resolve/main/isee/best_model.pt",
}


def checkpoint_is_valid(path):
    try:
        torch.load(path, map_location="cpu", weights_only=True)
        return True
    except Exception as exc:
        print(f"Existing checkpoint is not readable: {exc}")
        return False


for filename, url in model_files.items():
    path = Path(filename)
    needs_download = not path.exists()
    if filename == "best_model.pt" and path.exists():
        needs_download = not checkpoint_is_valid(path)

    if needs_download:
        if path.exists():
            path.unlink()
        print(f"Downloading {filename}...")
        urlretrieve(url, path)

    if filename == "best_model.pt" and not checkpoint_is_valid(path):
        raise RuntimeError(f"Downloaded {filename}, but it is still not a readable PyTorch checkpoint.")

    print(f"{filename}: {path.stat().st_size / 1024**2:.1f} MB")
```

## Export Case Data for WCS-Based Analysis

Save this run in an organized case folder under `examples/analysis_exports/`. The export includes the extrapolated magnetic field, physical coordinates, an HMI/SunPy reference map resampled to the model grid, and metadata. Later AIA cells save their raw and HMI-reprojected maps into the same case folder. `field_line_inclination.ipynb` reads the latest case automatically and uses the saved SunPy map WCS to create an inclination `Map`.

```python
b = mm.get_pred_from_numpy(model_input)
print(b.shape)

from pathlib import Path
import json
import re
import numpy as np
import astropy.units as u


def safe_path_part(value):
    return re.sub(r"[^A-Za-z0-9_.-]+", "_", str(value)).strip("_")


run_time = safe_path_part(d.replace(" ", "T").replace(":", ""))
try:
    active_region_id = f"noaa_{int(noaanum)}_harp_{int(harpnum)}"
except Exception:
    active_region_id = f"harp_{safe_path_part(harpnum)}"

case_dir = Path("examples/analysis_exports") / active_region_id / run_time
extrapolation_dir = case_dir / "extrapolation"
hmi_dir = case_dir / "maps" / "hmi"
metadata_dir = case_dir / "metadata"
for directory in (extrapolation_dir, hmi_dir, metadata_dir):
    directory.mkdir(parents=True, exist_ok=True)

export_path = extrapolation_dir / "rtmag_extrapolation.npz"
hmi_reference_path = hmi_dir / "hmi_reference_model_grid.fits"
metadata_path = metadata_dir / "case_metadata.json"
latest_case_path = Path("examples/analysis_exports/latest_case.txt")

# Match the model grid. SunPy maps use data shape (ny, nx), while the model
# field uses (nx, ny, nz, component), so inclination maps will use data.T.
hmi_reference_map = hmi_map.resample([nx, ny] * u.pixel)
hmi_reference_map.save(hmi_reference_path, overwrite=True)

# get_input() computes dx, dy, dz in Mm from the HMI plate scale,
# and sets z = np.arange(nz) * dz. Store the physical coordinates
# and the HMI reference map path for later SunPy/WCS use.
np.savez_compressed(
    export_path,
    b=b.astype(np.float32),
    x_coords=np.asarray(x, dtype=np.float32),
    y_coords=np.asarray(y, dtype=np.float32),
    z_coords=np.asarray(z, dtype=np.float32),
    dx_Mm=np.float32(dx),
    dy_Mm=np.float32(dy),
    dz_Mm=np.float32(dz),
    dx=np.float32(dx),
    dy=np.float32(dy),
    dz=np.float32(dz),
    component_order=np.array(["Bx", "By", "Bz"]),
    coordinate_units=np.array("Mm"),
    field_units=np.array("G"),
    case_dir=np.array(str(case_dir)),
    hmi_reference_map_path=np.array(str(hmi_reference_path)),
)

case_metadata = {
    "date": d,
    "harpnum": int(harpnum) if str(harpnum).isdigit() else str(harpnum),
    "noaanum": int(noaanum) if str(noaanum).isdigit() else str(noaanum),
    "nx": int(nx),
    "ny": int(ny),
    "nz": int(b.shape[2]),
    "dx_Mm": float(dx),
    "dy_Mm": float(dy),
    "dz_Mm": float(dz),
    "coordinate_units": "Mm",
    "field_units": "G",
    "extrapolation_path": str(export_path),
    "hmi_reference_map_path": str(hmi_reference_path),
}
metadata_path.write_text(json.dumps(case_metadata, indent=2))
latest_case_path.parent.mkdir(parents=True, exist_ok=True)
latest_case_path.write_text(str(case_dir))

print(f"Saved case folder: {case_dir}")
print(f"Saved extrapolation: {export_path}")
print(f"Saved HMI WCS reference map: {hmi_reference_path}")
print(f"Saved latest-case pointer: {latest_case_path}")
print(f"b shape: {b.shape} = (nx, ny, nz, 3)")
print(f"z range: {z[0]:.3f} to {z[-1]:.3f} Mm; dz = {dz:.3f} Mm")
```

## Export AIA Maps for This Case

Save the AIA maps used in the comparison plot. Raw AIA maps are saved under `maps/aia_raw/`, and the HMI-reprojected/resampled maps are saved under `maps/aia_reprojected/` so they share the model-grid WCS.

```python
from pathlib import Path
import json

aia_raw_dir = case_dir / "maps" / "aia_raw"
aia_reprojected_dir = case_dir / "maps" / "aia_reprojected"
aia_raw_dir.mkdir(parents=True, exist_ok=True)
aia_reprojected_dir.mkdir(parents=True, exist_ok=True)

aia_manifest = {}
for wavelength in [94, 131, 171, 193, 211, 304, 335]:
    raw_map = globals().get(f"aia{wavelength}_map")
    reprojected_map = globals().get(f"aia{wavelength}_map_repro")
    aia_manifest[str(wavelength)] = {}

    if raw_map is not None:
        raw_path = aia_raw_dir / f"aia_{wavelength}.fits"
        raw_map.save(raw_path, overwrite=True)
        aia_manifest[str(wavelength)]["raw"] = str(raw_path)

    if reprojected_map is not None:
        repro_path = aia_reprojected_dir / f"aia_{wavelength}_hmi_model_grid.fits"
        reprojected_map.save(repro_path, overwrite=True)
        aia_manifest[str(wavelength)]["reprojected"] = str(repro_path)

(metadata_dir / "aia_manifest.json").write_text(json.dumps(aia_manifest, indent=2))
print(f"Saved AIA maps to {aia_raw_dir} and {aia_reprojected_dir}")
```
